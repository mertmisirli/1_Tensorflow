```python
import pandas as pd
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.svm import LinearSVC
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=4)

pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('kmeans', kmeans)
])
```

```python
pipeline.fit(samples)
```

```python
labels = pipeline.predict(samples)
```

```python
labels
```

```python
df = pd.DataFrame({'labels' : labels, 'species' : species})
```

```python
ct = pd.crosstab(df['labels'], df['species'])
```

```python
ct
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Çapraz tabloyu çiz
plt.figure(figsize=(8, 6))
sns.heatmap(ct, annot=True, fmt="d", cmap="Blues")
plt.title("Cluster vs. True Label (Species)")
plt.xlabel("Gerçek Sınıf (species)")
plt.ylabel("Küme Etiketi (labels)")
plt.show()
```

```python
ct_percent = ct.div(ct.sum(axis=1), axis=0)

plt.figure(figsize=(8, 6))
sns.heatmap(ct_percent, annot=True, fmt=".2f", cmap="viridis")
plt.title("Kümelerdeki Gerçek Etiketlerin Oranı")
plt.xlabel("Gerçek Sınıf")
plt.ylabel("Küme")
plt.show()
```

```python
ct.plot(kind='bar', stacked=True)
plt.title("Küme Etiketlerine Göre Gerçek Etiket Dağılımı")
plt.xlabel("Küme Etiketi")
plt.ylabel("Sayı")
plt.legend(title="Gerçek Etiket")
plt.tight_layout()
plt.show()
```
